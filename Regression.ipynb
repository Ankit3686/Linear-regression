### Linear Regression
##### Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It provides valuable insights for prediction and data analysis. This article will explore its types, assumptions, implementation, advantages and evaluation metrics.

```python
#simple Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Sample Data
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([2, 4, 5, 4, 5, 6])

# Step 2: Create & Train the Model
model = LinearRegression()
model.fit(X, y)

# Step 3: Predict values
y_pred = model.predict(X)

# Step 4: Plot the Graph
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```

##### Ridge Regression

```python
#Ridge Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

```python
# Sample data (slightly noisy)
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([1.5, 2.0, 2.5, 3.7, 5.1, 6.0, 6.9, 7.8])

ridge_model= Ridge(alpha=1.0)
ridge_model
```

```python
ridge_model.fit(X,y)
#prediction 
predict=ridge_model.predict(X)
predict
```

```python
plt.scatter(X,y,color="red", label="Actual Data")
plt.plot(X,predict,color="blue", label="Regression Data")
plt.title("Ridge Regression")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid()
plt.show()
```

##### Random Forest Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
y = np.array([1.4, 1.9, 3.0, 3.8, 5.0, 6.1, 7.2, 7.9])
```

```python
model=RandomForestRegressor(n_estimators=100,random_state=0)
model.fit(X,y)
```

```python
x_test=np.arange(1.0, 8.1, 0.01).reshape(-1,1)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(x_test, y_predict, color='orange', label='Random Forest Prediction')
plt.grid()
plt.title("Random Forest Regression")
plt.xlabel("X values")
plt.ylabel("y values")
plt.legend()
plt.show()
```
